```python
import torch
import torchvision.models as models
from torchvision import transforms as T
import PIL.Image as Image
import cv2
from matplotlib import pyplot as plt

imgTensor = T.ToTensor()(Image.open('lena.jpeg'))
imgTensor = imgTensor.unsqueeze(0)
print(imgTensor.shape)
conv2d_1 = torch.nn.Conv2d(in_channels=3, out_channels=3, kernel_size=8)
maxpool = torch.nn.MaxPool2d(1)

# conv2d_2 = torch.nn.Conv2d(in_channels=4, out_channels=5, kernel_size=2)

aa = conv2d_1(imgTensor)
bb = maxpool(aa)
print(bb.shape)
bb = bb.squeeze(0)
bb = T.ToPILImage()(bb)
display(bb)
```
